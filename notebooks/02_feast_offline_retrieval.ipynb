# Feast offline retrieval

Requires Postgres (Feast online store) and `feast apply` from `feast_repo/` with `data/processed/account_daily_features.parquet` present.

```python
from datetime import datetime, timezone

import pandas as pd
from feast import FeatureStore

repo_path = "../feast_repo"
store = FeatureStore(repo_path=repo_path)

entity_df = pd.DataFrame(
    {
        "account_id": ["acc1"],
        "event_timestamp": [datetime(2024, 1, 3, tzinfo=timezone.utc)],
    }
)

training_df = store.get_historical_features(
    entity_df=entity_df,
    features=["account_daily_features:daily_tx_count", "account_daily_features:daily_amount_sum"],
).to_df()
training_df.head()
```
